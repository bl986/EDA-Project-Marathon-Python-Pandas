```python
# seasons for the data -> slower in summer than winter?

#spring 3-5, summer 6-8, fall 9-11, winter 12-2

#split between 2 decimals
```

```python
# the race_day looks like this: 02.02.2020, you just need the month which is the 2nd part, by index it's 1.
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head()
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11
                                            else 'Fall' if x > 8
                                            else 'Summer' if x > 5
                                            else 'Spring' if x > 2
                                            else 'Winter')
```

```python
df3.head(25)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
# 50mi only
```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
